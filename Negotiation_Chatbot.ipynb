```python
from transformers import pipeline

# Load sentiment analysis model from Hugging Face
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
import ipywidgets as widgets
from IPython.display import display

# Define the advanced sentiment negotiation chatbot function
class AdvancedSentimentNegotiation:
    def __init__(self):
        self.bot_price = 100
        self.current_offer = self.bot_price
        self.discount = 0

    def negotiate(self, user_input, user_message):
        # Analyze the sentiment of the user's message
        sentiment = sentiment_analyzer(user_message)[0]
        sentiment_label = sentiment['label']

        try:
            user_price = int(user_input)
        except ValueError:
            return "Please enter a numerical value in the offer box."

        # Check if the user accepted the offer
        if "accept" in user_message.lower() or "yes" in user_message.lower():
            if user_price == self.current_offer:
                return "Congratulations! You have accepted the offer. The negotiation is complete."
            else:
                return "Please enter the correct offer to accept."

        # Check if the user's offer is higher than the current offer
        if user_price >= self.current_offer:
            return "Congratulations! You have agreed on the price , The negotiation is complete."

        # Positive sentiment: User is polite, bot feels good and offers a discount
        if sentiment_label == "POSITIVE":
            if self.discount < 0.25:
                self.discount += 0.05  # Offer a 5% discount for being polite
                self.current_offer = self.bot_price * (1 - self.discount)
                return f"Because you're polite, the bot feels good and offers a {self.discount*100}% discount. New offer: ${self.current_offer:.2f} do you accept the offer."
            else:
                return f"The bot cannot offer a higher discount. The current offer is ${self.current_offer:.2f}."

        # Neutral sentiment: Bot continues with normal negotiation
        elif sentiment_label == "NEUTRAL":
            return negotiation_chatbot(user_input)

        # Negative sentiment: User is harsh, bot responds defensively or raises the price
        else:
            penalty = 0.1  # Increase the price by 10% for being harsh
            self.current_offer = self.bot_price * (1 + penalty)
            return f"The bot doesn't appreciate your tone. The price is now ${self.current_offer:.2f}. Would you like to make another offer?"

# Create the interface
interface = widgets.VBox([
    widgets.HTML(value=f"Initial Price: ${AdvancedSentimentNegotiation().bot_price}"),
    widgets.Text(value='', placeholder='Enter your offer', description='Offer:'),
    widgets.Text(value='', placeholder='Enter your message', description='Message:'),
    widgets.Button(description='Submit')
])

# Initialize the negotiation chatbot
negotiation_chatbot = AdvancedSentimentNegotiation()

# Define the callback function for the button
def submit_button_clicked(b):
    user_input = interface.children[1].value
    user_message = interface.children[2].value

    if not user_input.replace('.', '', 1).isdigit():
        print("Please enter a numerical value in the offer box.")
        interface.children[1].value = ''  # Clear the offer input
        interface.children[2].value = ''  # Clear the message input
        return

    if not user_message.replace(' ', '').isalpha():
        print("Please enter a string message.")
        interface.children[1].value = ''  # Clear the offer input
        interface.children[2].value = ''  # Clear the message input
        return

    response = negotiation_chatbot.negotiate(user_input, user_message)
    print(response)

    if "Congratulations" in response:
        interface.children[1].value = ''  # Clear the offer input
        interface.children[2].value = ''  # Clear the message input
    else:
        # Clear the input boxes
        interface.children[1].value = ''  # Clear the offer input
        interface.children[2].value = ''  # Clear the message input

# Link the button to the callback function
interface.children[3].on_click(submit_button_clicked)

# Display the interface
display(interface)
```
